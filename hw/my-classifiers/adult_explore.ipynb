```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("./adult/adult.data", header=None)
```

```python
data.head(10)
```

```python
data[14] = data[14].astype('category').cat.codes
```

```python
transformed_columns = []
for i in data.columns:
    col = data[i]
    if col.dtype == 'object':
        res = pd.get_dummies(col.astype('category'), prefix=str(i))
    else:
        res = col
    transformed_columns.append(res)

data_transformed = pd.concat(transformed_columns, axis=1)
data_transformed.head(10)
```
